## One hot encoding

- É a codificação de cada variável categórica em diferentes variáveis booleanas (também chamadas de variáveis dummy) que assumem valores 0 ou 1, indicando se uma categoria está ou não presente em uma observação.

- Usamos One-Hot Enconding quando precisamos estabelecer uma relação binária entre as categorias.

- <b>Limitação</b>: Expande a dimensão à medida que o número de colunas aumenta, o que pode levar a diversos problemas no processo de análise e modelagem preditiva.

###  Precisamos realmente aplicar One-Hot-Encoding em todas as categorias?

- Devemos determinar se a aplicação do One Hot encoding é realmente necessária em todas as colunas e também para todas as categorias. A aplicação dessa técnica em todas as colunas irá aumentar de forma significativa a dimensionalidade dos dados, podendo levar a problemas nas aplicações futuras. Neste exemplo, iremos aplicar essa técnica somente nas 10 maiores categorias, de cada uma das variáveis. É importante analisar se esse tipo de técnica irá ou não impactar os resultados de forma significativa. 

```python
#Linha de código para verificar o top 10 da variável X1
dados.X1.value_counts(ascending=False).head(10)
```

```python
# Criamos uma função para criar as novas variáveis com One-Hot-Encoding
def one_hot_top10(df, column, top_10):
    for i in top_10:
        df[column + "_" + i ] = np.where(dados[column]==i,1,0)
```

```python
for cols in dados.columns:
    top_10 = [x for x in dados[cols].value_counts().head(10).index]
    one_hot_top10(dados, cols, top_10)
```

```python
# Verificamos se algum valor nulo foi gerado
dados.isnull().any()
```
