## Label Encoding:

- Label Enconding (ou integer enconding) é a substituição de uma categoria por sua representação numérica correspondente.

- Usamos o label enconding quando temos um número baixo de categorias.

- <b>Limitação</b>: Label Enconding não é adequado para modelos lineares como regressão logística.

```python
import pandas as pd
import datetime
```

```python
data_atual = datetime.datetime.now()
```

```python
data_atual
```

```python
#Gerando uma lista de datas com dias anteriores ao dia de hoje
data_list = [data_atual-datetime.timedelta(days=x) for x in range(0,20)]
```

```python
data_list
```

```python
#Transformando a lista de datas num dataframe
df = pd.DataFrame(data_list)
```

```python
df.head()
```

```python
# Ajusta o nome da coluna
df.columns = ['data']
```

```python
df.head()
```

```python
df['dia_da_semana'] = df['data'].dt.day_name()
```

```python
df
```

```python
dict_dia_semana = {'Monday':1, 
                   "Tuesday":2,
                   "Wednesday":3,
                   "Thursday":4,
                   "Friday":5,
                   "Saturday":6,
                   "Sunday":7}
```

```python
df['dia_semana_cod'] = df['dia_da_semana'].map(dict_dia_semana)
```
